```python
#1)Загрузить данные в pandas
import pandas as pd

df_data = pd.read_csv('u.data', sep='\t', header=None)
df_genre = pd.read_csv('u.genre', sep='|', header=None)
df_info = pd.read_csv('u.info', sep=' ', header=None)
df_item = pd.read_csv('u.item', sep='|', encoding='latin_1', header=None)
df_occup = pd.read_csv('u.occupation', header=None)
```

```python
#-------------------------------------------------------------------------------------------------------------
#2)Посчитайте среднюю оценку, которую поставил каждый пользователь
df_data.columns = ['user id', 'item id', 'rating', 'timestamp']
df_data.head()
```

```python
df_data.groupby('user id').rating.mean()
```

```python
#-----------------------------------------------------------------------------------------------------------------
#3)Отсортируйте фильмы по количеству оценок
df_data.groupby('item id').rating.count().sort_values(ascending=False)
```

```python
#----------------------------------------------------------------------------------------------------------------------
#4)Средствами pandasa соедините данные по оценкам с данными по пользователям (u.user)
df_user = pd.read_csv('u.user', sep='|', header=None)
df_user.columns = ['user id', 'age', 'gender', 'occupation', 'zip code']
df_users_ratings = pd.merge(df_data, df_user, on='user id', how='inner')
df_users_ratings.head()
```

```python
#---------------------------------------------------------------------------------------------------------------
#5)Построить модель, которая предсказывает оценку для фильма и пользователя (воспользоваться RandomForestRegressor, тк мы решаем задачу регрессии)
df_item.columns = ['item id', 'movie title','release date','video release date',
              'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation',
              'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical','Mystery','Romance', 'Sci-Fi',
              'Thriller', 'War','Western']
df_item.groupby('video release date').head()
df_item[df_item['video release date'].notnull()]
```

```python
df_items = df_item[df_item.columns.difference(['video release date', 'IMDb URL'])]
df_items.head()
```

```python
df_users_ratings_items = pd.merge(df_items, df_users_ratings, on='item id', how='inner')
df_occup.columns = ['occupation']
df_occup['occupation_index'] = df_occup.index
df_users_ratings_items = pd.merge(df_users_ratings_items, df_occup, on='occupation', how='inner')
```

```python
df_users_ratings_items = df_users_ratings_items[df_users_ratings_items.columns.difference(['movie title', 'occupation', 'release date','timestamp', 'zip code'])]
df_users_ratings_items['gender'].replace('F', 0, inplace=True)
df_users_ratings_items['gender'].replace('M', 1, inplace=True)
df_users_ratings_items.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df_users_ratings_items[df_users_ratings_items.columns.difference(['rating'])]
X.head()
```

```python
y = df_users_ratings_items['rating']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
#-----------------------------------------------------------------------------------------------------------------------
#6)Оценить качество регрессора на отложенной (train_test_split) выборке по одной из метрик на выбор - RMSE, RMAE, R^2
from sklearn.metrics import mean_squared_error
from math import sqrt

prediction = model.predict(X_test)
MSE =mean_squared_error(y_test, prediction)
print("MSE:", MSE)
RMSE = sqrt(MSE)
print("RMSE:", RMSE)
```
